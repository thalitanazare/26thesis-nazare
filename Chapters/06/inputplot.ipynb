```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# ============================================================
# Global font settings
# ============================================================

plt.rcParams.update({
    "font.family": "Times New Roman",
    "font.size": 24
})


# ============================================================
# List of sea-state files and corresponding indices
# ============================================================

sea_state_files = [
    ("data1_1p2Hs_10Tp.csv", 1),
    ("data4_6p2Hs_10Tp.csv", 2),
    ("data5_1p2Hs_8Tp.csv", 3),
    ("data8_6p2Hs_8Tp.csv", 4)
]


# ============================================================
# Data range to be analyzed
# ============================================================

start_index = 5000
end_index = 6000
segment_length = end_index - start_index

# Time vector for plotting
time = np.arange(segment_length)


# ============================================================
# Generate excitation force plots for each sea state
# ============================================================

for file_name, state_idx in sea_state_files:
    # Load the CSV file
    data = pd.read_csv(file_name, header=None)

    # Extract the excitation force signal (column 0)
    u = data.iloc[start_index:end_index, 0].values

    # Create the figure
    plt.figure(figsize=(20, 5))
    plt.plot(time, u, "b")

    # Axis labels
    plt.xlabel("Time")
    plt.ylabel(r"$u$")

    # Optional title
    # plt.title(f"Excitation Force - Sea State {state_idx}")

    # Add grid
    plt.grid(True)

    # Save the figure
    plt.savefig(f"u_input_ss{state_idx}.svg", format="svg")

    # Close the figure to free memory
    plt.close()
```
